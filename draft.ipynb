```python
from numpy.linalg import svd
a = np.random.randn(9, 6) + 1j*np.random.randn(9, 6)
u, s, vt = np.linalg.svd(a, full_matrices=False)
C0 = u.dot(np.sqrt(np.diag(s))).T
W0 = np.sqrt(np.diag(s)).dot(vt)
np.allclose(a, C0.T@W0)
```
